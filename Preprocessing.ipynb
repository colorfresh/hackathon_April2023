```python
# Category lists
time_related = ['time', 'wait', 'forever', 'slow', 'fast', 
                'takes forever', 'busy', 'crowded', 'line',
                'delay', 'delayed' 'wasted', 'late']

service_related = ['rude', 'bad', 'horrible', 'management',
                   'manager', 'customer service', 'service',
                   'arrogant', 'unfriendly', 'terrible', 'mistake',
                   'unfortunate', 'pathetic', 'awful']

pattern_1 = re.compile('|'.join(time_related))
pattern_2 = re.compile('|'.join(service_related))

df_merged['is_time'] = df_merged['text'].str.contains(pattern_1, regex=True).apply(lambda x : 1 if x else 0)
df_merged['is_service'] = df_merged['text'].str.contains(pattern_2, regex=True).apply(lambda x : 1 if x else 0)

df_merged.head()
```

```python
df_cleaned = df_merged[df_merged['Restaurants'] == 1]

# Subset the data frame to include relevant columns
df_subset = df_cleaned[['stars_Biz', 'is_service', 'is_time']]

# Calculate the count of reviews for each combination of 'is_service' and 'is_time' values and 'stars_Biz' values
grouped_data = df_subset.groupby(['is_service', 'is_time', 'stars_Biz']).size().reset_index(name='count')

# Pivot the data to create the desired matrix
matrix_data = grouped_data.pivot(index=['is_service', 'is_time'], columns='stars_Biz', values='count')

# Rename the columns with the stars_Biz values
matrix_data.columns = ['1 Star', '2 Stars', '3 Stars', '4 Stars', '5 Stars']

# Create the crosstab matrix
matrix = pd.DataFrame(matrix_data.to_records())

# Reorder the columns
matrix = matrix[['is_service', 'is_time', '1 Star', '2 Stars', '3 Stars', '4 Stars', '5 Stars']]

matrix
```
