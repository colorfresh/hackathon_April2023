```python
df_biz = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SoCalRHackathon/CA_Biz.csv")


import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_word(word, pos_tag):
    pos = get_wordnet_pos(pos_tag)
    if pos:
        return lemmatizer.lemmatize(word, pos)
    else:
        return lemmatizer.lemmatize(word)

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

def lemmatize_city(city):
    tokens = nltk.word_tokenize(city)
    pos_tags = nltk.pos_tag(tokens)
    
    lemmatized_tokens = [lemmatize_word(token, pos_tag) for token, pos_tag in pos_tags]
    lemmatized_city = ' '.join(lemmatized_tokens)
    
    return lemmatized_city


# df_biz['city'] = df_biz['city'].str.lower()
df_biz['Lemmatized_City'] = df_biz['city'].apply(lemmatize_city)


df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('santa Barbara', 'Santa Barbara')
df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('Santa Barbara ,', 'Santa Barbara')
df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('SANTA BARBARA AP', 'Santa Barbara')
df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('Santa Barbara & Ventura Counties', 'Santa Barbara')
df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('Santa Barbra', 'Santa Barbara')
df_biz['Lemmatized_City'] = df_biz['Lemmatized_City'].str.replace('SANTA BARBARA', 'Santa Barbara')


df_biz # ['Lemmatized_City']
```

```python
# geojson heatmap

import pandas as pd
import folium
from geopy.geocoders import Nominatim

# # create a sample dataframe with California cities and counts
# data = {
#     'City': ['Los Angeles', 'San Diego', 'San Francisco', 'Sacramento', 'San Jose'],
#     'Count': [100, 50, 75, 30, 80]
# }
# df = pd.DataFrame(data)

# create a geocoder object
geolocator = Nominatim(user_agent='my-app', timeout=10)

# add latitude and longitude coordinates for each city
df_biz['Location'] = df_biz['Lemmatized_City'] + ', CA'


```

```python

from geopy.exc import GeocoderTimedOut

def geocode_location(location, geolocator, timeout=10):
    try:
        return geolocator.geocode(location, timeout=timeout)
    except GeocoderTimedOut:
        print(f'Geocoding request timed out for location "{location}"')
        return None

geolocator = Nominatim(user_agent='my-app', timeout=10)

df_biz['Latitude'] = None
df_biz['Longitude'] = None

for i, row in df_biz.iterrows():
    location = row['Location']
    result = geocode_location(location, geolocator)
    if result is not None:
        df_biz.at[i, 'Latitude'] = result.latitude
        df_biz.at[i, 'Longitude'] = result.longitude
    else:
        print(f'Geocoding failed for location "{location}"')


df_biz
```

```python
df_biz['Latitude'] = df_biz['Location'].apply(lambda x: geolocator.geocode(x).latitude)
df_biz['Longitude'] = df_biz['Location'].apply(lambda x: geolocator.geocode(x).longitude)

# create a map centered on California
m = folium.Map(location=[37.7749, -122.4194], zoom_start=6)

# add markers for each city, scaled by count
for index, row in df_biz.iterrows():
    folium.Circle(
        location=[row['Latitude'], row['Longitude']],
        radius=row['review_count']*100,
        popup=row['Lemmatized_City'] + ': ' + str(row['review_count']),
        color='crimson',
        fill=True,
        fill_color='crimson'
    ).add_to(m)

# display the map
m
```
